Line plot : The line plot is one of the most basic plot in seaborn library.  This plot is mainly used to visualize the data in form of some time series.

```python
import seaborn as sns
  
  
sns.set(style="dark")
fmri = sns.load_dataset("fmri")
  
# Plot the responses for different\
# events and regions
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             style="event",
             data=fmri)
```
